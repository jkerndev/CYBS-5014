## KAPE Live Response Data Acquisition

Captures image of C drive (default on the Virtual Machine this notebook is executing on) based on targets:
- Event Logs (Windows EVTX, incl. Security/System/Application and common Microsoft-* channels).
- Evidence of Execution (prefetch, shimcache, Amcache/RecentFileCache/UserAssist).
- Registry (System & User) for configs, MRU, autoruns/persistence (Run keys, services, tasks).
- File System Metadata for MFT, UsnJrnl, Boot logs, etc.

and modules:
- Parses the MFT to create file-activity timelines (Event Logs, MFT).
- Registry-hive parser.
- Running processes, PIDs, and command lines.
- Summarizes system build, uptime, and installed patches.
- Parses EVTX into human-readable format.

```python
KAPE = r"C:\standalone\kape\kape.exe"
OUTPUT = r"C:\Users\vboxuser\repos\CYBS-5014\lab4\output\kape-targets"
TARGETS = "EventLogs,EvidenceOfExecution,RegistryHives,FileSystem"

print(f"Command to be run: \n ---- \n {CMD} \n ---- \n Running... \n")

CMD = f"{KAPE} --tsource C: --tdest {OUTPUT} --tflush --target {TARGETS}"

! $CMD
```

```python
KAPE = r"C:\standalone\kape\kape.exe"
OUTPUT = r"C:\Users\vboxuser\repos\CYBS-5014\lab4\output\kape-modules"
MODULES = "MFTECmd,RegRipper,LiveResponse_ProcessDetails,SystemInformation,Events-Ripper"

print(f"Command to be run: \n ---- \n {CMD} \n ---- \n Running... \n")

CMD = f"{KAPE} --msource C: --mdest {OUTPUT} --mflush --module {MODULES} --mef json"

! $CMD
```
